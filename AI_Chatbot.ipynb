```python
print("ChatBot: Hello! I'm your AI assistant.")
print(" ChatBot: You can chat with me. Type 'exit' anytime to end the conversation.\n")

responses = {
    "hello": "Hello! Nice to meet you ",
    "hi": "Hi! How are you today?",
    "how are you": "I'm doing great! Thanks for asking. How about you?",
    "what is your name": "My name is ChatBot. I'm a rule-based AI assistant.",
    "who created you": "I was created using Python and simple AI rules.",
    "help": "You can greet me, ask my name, or simply chat with me.",
    "bye": "Goodbye! Have a wonderful day ",
    "good morning": "Good morning! Hope you have an amazing day ahead.",
    "good evening": "Good evening! How was your day?",
    "thank you": "You're welcome! Happy to help."
}

while True:
    user_input = input("You: ").lower().strip()

    if user_input == "exit":
        print(" ChatBot: It was nice talking to you. Goodbye!")
        break

    response = responses.get(
        user_input,
        " ChatBot: Sorry, I didn't understand that. Can you try asking something else?"
    )

    print(response)
```
